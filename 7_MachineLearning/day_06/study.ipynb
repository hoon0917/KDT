### 데이터 탐색적 분석 시 순서
 1. 전처리 : 
  - 모든 것을 수치화 : 범주형 -> 인코딩
 2. 지도/비지도/강화 중 학습 방법 선택
 3. 학습 알고리즘 선택
  - 독립변수와 종속변수(fearuter와 label)의 관계가 선형인지 비선형인지 파악 후 모델 선정

### 선형 모델
 1. 피쳐와 타겟의 관계가 선형관계
 2. 정규분포 기반
 3. 선형식 : y=ax+b => wx+b => 
  - 우리가 사용하는 식 : W0X+W1X+W2X+W3X+W4X+...+WnX
 4. Coefficent(Weight) 사용 이유 : 학습을 통해서 다양한 피쳐들의 영향정도를 정확하게 반영하기 위해
  - Coefficient는 회귀 모델에서 각 독립 변수(피처)에 대해 계산된 값. 이 값은 해당 피처가 종속 변수(label)에 미치는 영향  
※ 학습 fit은 최적의 coefficent를 찾아가는 과정
 5. 일반화를 목적으로 피쳐를 줄이는 작업을 하는데 피쳐를 줄이다보니까 자꾸 과대적합하는 문제가 발생  
    그래서 과대적합 방지를 위한 방법으로 가중치에 규제(릿지,랏소,엘라스틱넷)를 가함
    - 릿지 : 가중치를 0에 가깝게
    - 랏쏘 : 가중치를 0으로 
    - 엘라스틱넷 : 릿지 + 랏쏘

#### OvR / OvA (One vs Rest / One vs All) 분류  

[0,1,2,3,4] 5개 분류 -> 2진 분류기 5개  

0 -> 0 vs [1,2,3,4] -> 2진분류 -> 0일 확률  
1 -> 1 vs [0,2,3,4] -> 2진분류 -> 1일 확률  
2 -> 2 vs [1,0,3,4] -> 2진분류 -> 2일 확률  
3 -> 3 vs [1,2,0,4] -> 2진분류 -> 3일 확률  
4 -> 4 vs [1,2,3,0] -> 2진분류 -> 4일 확률
