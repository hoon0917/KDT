- [1] 모듈로딩 및 데이터 준비

```python
# 모듈 로딩
from urllib.request import urlretrieve
```

```python
## 데이터 준비

DATA_FILE = '../data/2016-10-20.txt'
```

- [2] 말뭉치 생성

```python
# 모듈로딩
from soynlp.utils import DoublespaceLineCorpus
```

```python
# 파일에서 데이터 말뭉치 => 코퍼스 생성
data_corpus = DoublespaceLineCorpus(DATA_FILE)
```

```python
print(f'{type(data_corpus)}, 문서 수 : {len(data_corpus)}')
```

```python
# 5개 문서 추출(말뭉치 생성)
for idx, doc in enumerate(data_corpus) :
    if idx == 5 :break
    print(f'[{idx}]====\n{doc}')
```

- [3] 단어 추출

```python
# 단어 추출 모듈 로딩
from soynlp.word import WordExtractor
```

```python
# 단어 추출 인스턴스 생성
wordExt = WordExtractor()
```

```python
# 텍스트 데이터 제공 비지도학습 진행
wordExt.train(data_corpus)
```

```python
# 단어 추출
## Cohesion 문맥 뒤에 특정 단어가 올 확률을 예측하는 것
## Branching Entropy : 하나의 단어를 중간에 끊으면 다음 올 글자 예측이 쉬워짐
## ex) 대한민국에서 대한과 민국을 자르면 대한민국이라는 단어를 예측하기 쉬움
wordExt.extract()
```
