#### [Matplotlib 활용한 시각화]
- matplotlib.pyplot 모듈 사용
- 다양한 그래프 관련 함수 & 클래스 존재

```python
# [1] 모듈 로딩
# 패키지명.모듈명 
import matplotlib.pyplot as plt
```

```python
# [2] 그래프 그리기 - 선(Line) 그래프
# plot(y축데이터 기본)
plt.plot([1,2,3,4])
plt.xlabel('Data')
plt.ylabel('Result')
plt.show()
# 객체를 변수명 설정해서 저장하는 이유는 관련 데이터들을 가져오기 위해서 ex) .get()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(x축데이터, y축데이터)
plt.plot([10,20,30],[5,10,15])
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(x축데이터, y축데이터, fmf(포맷스트링))
# fmf - 포맷스트링 : 그래프 색상, 두께, 스타일 등 지정
plt.plot([10,20,30],[5,10,15], 'r--')
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(x축데이터, y축데이터, fmf(포맷스트링))
# fmf - 포맷스트링 : 그래프 색상, 두께, 스타일 등 지정
plt.plot([10,20,30],[5,10,15], 'r*-')
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(x축데이터, y축데이터, line2D 속성으로 설정) 
# fmf - 포맷스트링 : 그래프 색상, 두께, 스타일 등 지정
# # color='red', marker='o',
#          linestyle='dashed', markersize=10,
#          linewidth=4
# - mec(markeredgecolor) => 'blue'
# - mew(makeredgewidth)
plt.plot([10,20,30],[5,10,15], color='red', marker='o',
         linestyle='dashed', markersize=10,
         linewidth=4, mec='blue',mew=4)
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plt.plot([10,20,30],[15,25,35],'bo--')
# plt.plot([10,20,30],[18,20,35],'ro--')
# plt.plot([10,20,30],[20,20,20],'ko-')
plt.plot([10,20,30],[18,20,35],'ro--',[10,20,30],[15,25,35],'bo--',[10,20,30],[20,20,20],'ko-')
plt.show()
```

```python
# [3] 그래프의 부가적인 기능 추가
# - 그래프 제목
# - x축, y축 label
# 범례 => 그래프의 label속성값을 범례에서 사용함
plt.plot([10,20,30],[10,25,5],'bo--',label='V')
plt.plot([10,20,30],[13,9,27],'ro--',label='2')
plt.title('[MPLITB LINE]')
plt.xlabel('DATA')
plt.ylabel('RESULT')
plt.legend()
plt.show()
```

```python
# [3] 그래프의 부가적인 기능 추가
# - 그래프 제목
# - x축, y축 label
# 범례 => 그래프의 label속성값을 범례에서 사용함
#         loc속성으로 위치 지정 가능 [기본값:best]
plt.plot([10,20,30],[10,25,5],'bo--',label='V')
plt.plot([10,20,30],[13,9,27],'ro--',label='2')
plt.title('[MPLITB LINE]')
plt.xlabel('DATA')
plt.ylabel('RESULT')
plt.legend(loc='center')
plt.show()
```
