##### [ Matplotlib 활용 데이터 시각화]
 - X축과 Y축 다양한 처리

```python
# [1] 모듈 로딩
from matplotlib import pyplot as plt 
```

```python
# [2] 그래프 그리기 - 기본
y_data = [1,3,5,7,9]
x_data=['2024-01-01','2024-05-01','2024-07-01','2024-09-01','2024-11-01']
plt.plot(x_data,y_data,marker='o')
plt.show()
```

```python
# [2] 그래프 그리기 - X축과 Y축 범위 설정
y_data = [1,3,5,7,9]
x_data=['2024-01-01','2024-05-01','2024-07-01','2024-09-01','2024-11-01']
plt.plot(x_data,y_data,marker='o')
plt.ylim(bottom=0,top=15)
a,b=
plt.show()
```

```python
# [2] 그래프 그리기 - X축과 Y축의 눈금 설정
y_data = [1,3,5,7,9]
x_data=['2024-01-01','2024-05-01','2024-07-01','2024-09-01','2024-11-01']
plt.plot(x_data,y_data,marker='o')
plt.ylim(bottom=0,top=15)
plt.xticks(rotation=50)
plt.show()
```

```python
# [2] 그래프 그리기 - 그래프 뜨는 사이즈 지정
y_data = [1,3,5,7,9]
x_data=['2024-01-01','2024-05-01','2024-07-01','2024-09-01','2024-11-01']

plt.figure(figsize=(10,5)) # 원하는 창의 크기 설정
plt.plot(x_data,y_data,marker='o')
plt.ylim(bottom=0,top=15)
plt.xticks(rotation=50)
plt.show()
```

```python
# [2] 그래프 그리기 - 그래프 출력 창 figure를 자동 맞춤
y_data = [1,3,5,7,9]
x_data=['2024-01-01','2024-05-01','2024-07-01','2024-09-01','2024-11-01']


plt.plot(x_data,y_data,marker='o')
plt.ylim(bottom=0,top=15)
plt.xticks(rotation=50)
plt.tight_layout()   # 그래프 출력 창 사이즈 알아서 조절 가장 tight 하게!!!!!!!
plt.show()
```

```python
# [2] 그래프 그리기 - 눈금(ticks) 설정
y_data = [1,3,5,7,9]
x_data=['2024-01-01','2024-05-01','2024-07-01','2024-09-01','2024-11-01']


plt.plot(y_data,marker='o')
plt.ylim(bottom=0,top=10)
plt.xlim(right=7,left=0)
font_config={'fontsize':22,'fontweight':'bold'}
plt.xticks([-10,2,4,6,8],['A','B','C','D','E'],**font_config)
plt.show()
# font-config 라는 Dict를 하나 만들어서 가져오기 편하게 할 수 있음
```
