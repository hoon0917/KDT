#### [ Matplotlib 활용 시각화 - 막대 그래프 ]
- 가장 많이 사용되는 그래프
- 데이터를 비교할 때 사용

```python
# [1] 모듈로딩
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# [2] 데이터 준비
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

```

```python
# [3] 데이터 시각화 - 수직 막대그래프
plt.bar(fruits,counts, width=0.3, color=bar_colors,label=bar_labels)
plt.legend()
plt.show()
```

```python
# [3] 데이터 시각화 - 수평 막대그래프
plt.barh(fruits,counts, height=0.5, color=bar_colors,label=bar_labels)
plt.legend()
plt.show()
```

```python
# [3] 데이터 시각화 - 수직 막대그래프 + 선그래프
plt.bar(fruits,counts, width=0.3, color=bar_colors,label=bar_labels)
plt.plot(counts,'bo--')
plt.legend()
plt.show()
```
