```python
# [1-1] 모듈 로딩
import pandas as pd                     # 데이터 분석용 모듈
import matplotlib.pyplot as plt         # 데이터 시각화용 모듈
```

```python
# [1-2] 데이터 준비
filedata=r'C:\Users\KDP-17\EX_PANDAS6\data\시도별 전출입 인구수.xlsx'
```

```python
# [1-3] DATA ==> DataFrame 형태 저장
dataDF=pd.read_excel(filedata,engine='openpyxl')
```

```python
### 기본 정보 확인 후 결과
# - 불필요한 행/컬럼 : 0번행 삭제
# - 결측치 : '전출지별' 컬럼 ==> 이전 행 값으로 채우기
# - 이상치 : 행별 '-'값 수치값과 섞여 있음 ==> 0 처리 필요
```

- [3] 데이터 전처리 
    * 기본확인 파악된 내용들 처리
    * 불필요 데이터 삭제 : 0번행 삭제
    * 결측치 데이터 처리 :'전출지별' 컬럼 처리
    * 이상치 데이터 처리 : '-'에 대한 처리
    * 데이터 자료형 일치 : 형변환

```python
# [3-1] 불필요 데이터 처리
copyDF=dataDF.drop(0)
print(f'확인 \n{copyDF.head(5)}')
```

```python
# [3-2] 결측치 데이터 처리 => isna()/isnull()
copyDF['전출지별']=copyDF['전출지별'].ffill()
```

```python
# - 1984,1985,1986 컬럼에서 결측치 데이터만 추출
copyDF.columns
dataDF.isna().sum()
a=copyDF[[1984,1985,1986]].isna() 
a[a.sum(axis=1)>0]
```

```python
copyDF[[1984,1985,1986]]=copyDF[[1984,1985,1986]].fillna(0)
copyDF.info()
```

```python
# [3-3] 이상치 데이터 처리
copyDF.replace('-',0,inplace=True)
```

```python
# [3-4] 자료형 변환 : 48개의 컬럼 int로 형변환
# zip 함수 내 ',' 뒤는 48개를 만드는 숫자
d_=dict(zip(copyDF.columns[2:].to_list(),['int']*(copyDF.shape[1]-2)))
# 방법2 dict(zip(copyDF.columns[2:].to_list(),['int']*len(copyDF.columns[2:].to_list())))
copyDF=copyDF.astype(d_)
```

- [4] 데이터 가공
    * 대구에서 전지역 전입 인구수 분석
    * 전출지별 - '대구광역시'
    * 전입지별 - '대구광역시','전국'제외

```python
# - 대구 광역시 데이터만 추출
mask=(copyDF['전출지별']=='대구광역시')
cityDF=copyDF[mask]
cityDF
```

```python
# - 전출지별 컬럼 삭제
cityDF=cityDF.drop('전출지별',axis=1)

# - 전입지별 컬럼 행 인덱스 설정
cityDF=cityDF.set_index('전입지별')
cityDF
```

```python
# - 대구광역시, 전국 행 삭제
cityDF.drop(['대구광역시','전국'],inplace=True)
cityDF
```

- [5] 분석 : 대구에서 전지역 전입 인구수 분석

```python
# [5-1] 폰트 적용
from matplotlib import font_manager as fm
from matplotlib import rc

# 적용할 폰트 파일
font_file = r'C:\Windows\Fonts\malgun.ttf'

# 폰트 패밀리 이름 가져오기
font_name=fm.FontProperties(fname=font_file).get_name()

# 새로운 폰트 패밀리 이름 지정
rc('font',family=font_name)
```

```python
# [5-2] 광역시 및 특별시와 일반 도 구분하고 시각화 하기
print(f'광역시 및 특별시 :{cityDF.index[:7]}')
print(f'광역시 및 특별시 :{cityDF.index[7:]}')
fig=plt.figure(figsize=(15,7))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)
for i in cityDF.index[:7] :
    country=[i]
    value=cityDF.loc[i]
    ax1.bar(country,value)
    ax1.set(title='대구 전출 인구')
for i in cityDF.index[7:] :
    country=[i]
    value=cityDF.loc[i]
    ax2.bar(country,value)
    ax2.set(title='대구 전출 인구')
    plt.tight_layout()


```
