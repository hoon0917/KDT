- 컬럼(변수)의 데이터 분표 비율 시각화

```python
# [1] 모듈 로딩
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from matplotlib import font_manager as fm
from matplotlib import rc

# 적용할 폰트 파일
font_file = r'C:\Windows\Fonts\malgun.ttf'

# 폰트 패밀리 이름 가져오기
font_name=fm.FontProperties(fname=font_file).get_name()

# 새로운 폰트 패밀리 이름 지정
rc('font',family=font_name)
```

```python
# [2] 데이터 준비
file_name=r'C:\Users\KDP-17\EX_PANDAS6\data\auto_mpg.csv'
```

```python
# [3] csv ==> DataFrame 저장
dataDF=pd.read_csv(file_name)
dataDF
```

```python
# - 컬럼별 타입 확인
dataDF.info()
```

```python
# [4] 특정 컬럼의 데이터 비율
# - 자동차 생산국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
# - 자동차 생산국가에 대한 비율
round(dataDF['origin'].value_counts()/dataDF['origin'].count(),2)
```

```python
originSR=dataDF['origin'].value_counts()
x=originSR.index.to_list()
h=originSR.values.tolist()
print(type(x),type(h))
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - bar 그래프
originBar=plt.bar(x,h,color=['red','yellow','green'])

# 막대에 눈금 설정
plt.bar_label(originBar,h)

# x축 눈금 설정
plt.xticks([1,2,3],['미국','유럽','일본']) # [1,2,3]은 눈금이 찍히는 위치값 뒤에는 라벨

plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h,labels=['US','EU','JP'],autopct='%.1f%%',counterclock=True)
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h,labels=['US','EU','JP'],autopct='%.1f%%',startangle=90)
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h,labels=['US','EU','JP'],autopct='%.1f%%',explode=[0,0,0.1])
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h,labels=['US','EU','JP'],autopct='%.1f%%',
        wedgeprops={'linewidth':2,'edgecolor':'w','width':0.6})
plt.show()
```
