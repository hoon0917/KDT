#### [ 시도별 전출입 인구 시각화]
- 대구에서 타지역으로 이동한 인구 시각화

```python
# [모듈 로딩]
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# [자료 준비]
file_name=r'C:\Users\KDP-17\EX_PANDAS6\data\시도별 전출입 인구수.xlsx'
```

```python
# xlsx => DF
dataDF=pd.read_excel(file_name)
dataDF.head(5)
```

```python
# 데이터 전처리
# - 0번 행 삭제
dataDF=dataDF.drop(0,axis=0)
```

```python
dataDF
```

```python
# 데이터 정보 확인
dataDF.info()
```

```python
# 결측치 확인
dataDF.isna().sum()
```

```python
# '-'값 0으로 치환 
dataDF.replace('-',0,inplace=True)
```

```python
# nan값 치환
dataDF['전출지별']=dataDF['전출지별'].ffill()
```

```python
# 전출지 기준 멀티인덱싱
dataDF=dataDF.set_index(['전출지별','전입지별'])
dataDF.head()
```

```python
# 대구광역시 데이터 가져오기
dataDF_Daegu=dataDF.loc['대구광역시']
```

```python
# 한글 폰트 설정 => 폰트 매니저 모듈 사용
from matplotlib import font_manager as fm
from matplotlib import rc

# 적용할 폰트 파일
font_file = r'C:\Windows\Fonts\malgun.ttf'

# 폰트 패밀리 이름 가져오기
font_name=fm.FontProperties(fname=font_file).get_name()

# 새로운 폰트 패밀리 이름 지정
rc('font',family=font_name)
```

```python
# 시각화 하기
for i in dataDF_Daegu.index:
    plt.plot(dataDF_Daegu.loc[i])
    plt.title('대구 연도별 전출 인구')
    plt.ylabel('전출 인구수')
    plt.xlabel('연도')
```
