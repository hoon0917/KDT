boolean indexing
sr/df에서 원하는 조건의 데이터만 추출하기 위한 것
True/False로 구성된 데이터를 인덱스로 사용

```python
import pandas as pd
```

```python
data={'name':['마징가','배트맨','홍길동'],
              'Age':[45,19,65],
              'gender':['남','남','여']}
Df=pd.DataFrame(data)
Df
```

```python
idx=Df['gender'] =='남'
Df[idx]
```

```python
idx2=(Df['gender'] == '남') & (Df['Age']>= 20)
Df[idx2]
```
