## [그룹화 처리]
- 주제 : 타이타닉에서 선실의 등급에 따른 생존 비율을 분석

```python
import pandas as pd
```

```python
file='titanic.csv'
```

```python
dataDF=pd.read_csv(file)
dataDF
```

```python
# 분석 주제 타이타닉 선실의 등급에 따른 생존 비율 분석
# 필요컬럼 : survived, pclass
suvDF=dataDF[['survived','pclass']]
suvDF
```

```python
# 전처리1 nan값 확인
suvDF.isna().sum()
```

```python
# 그룹화
sub_obj= suvDF.groupby('pclass')
sub_obj
```

```python
# 그룹의 키 및 값 확인
sub_obj.groups
print(sub_obj.groups.keys())
print(sub_obj.groups.values())
```

```python
# 살아남은 비율 : 각 등급의 살아남은 사람 / 각 등급의 전체사람
```

```python
# 각 등급의 전체 사람 => sub_obj.count()
sub_obj.count()
```

```python
# 각 등급일때 살아남은 사람의 합 구하기 => sub_obj.sum()
sub_obj.sum()
```

```python
sub_obj.sum()/sub_obj.count()*100
```

```python
dataDF.survived.count() - dataDF.survived.sum()
```

```python
sub_obj.agg(['count','sum'])
```

 - 사용자 정의 함수 지정하기

```python
def myfunc(obj):
    result=obj.count()-obj.sum()
    return result
```

```python
sub_obj.agg(['count','sum',myfunc])
```

```python
dataDF2=dataDF.groupby('pclass')['survived'].agg(['count','sum',myfunc])
```

```python
dataDF2.rename(columns={'myfunc':'비율'})
```
