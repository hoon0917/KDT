### DataFrame에서 행/열 추가
- 행 방향 추가
- 열 방향 추가

```python
# 모듈 로딩
import pandas as pd
```

```python
# 데이터 준비
data={'name':['홍길동','마징가','원더우먼'],
      'job':['의적','로봇','영웅'],
      'gender':['남','남','여'],
      'age':[10,20,30],
      'height':[180,300,200]}
```

```python
dataDF=pd.DataFrame(data)
dataDF
```

[실습] 행추가 : 변수명.loc['추가할 행이름'] = 데이터
* 변수명.iloc[] X
* 데이터수 == 컬럼수 겨웅 행 추가 가능
* 데이터수 == 1 경우도 행 추가 가능
* 그 외의 경우 행 추가 X

```python
## 1개 행 추가 => 데이터 수가 열 개수와 동일한 경우
dataDF.loc[3]=['배트맨','박쥐','남',40,190]
dataDF
```

```python
## 1개 행 추가 => 데이터 수가 1개 => 모든 컬럼에 동일한 값 저장
dataDF.loc[5]=100
dataDF
```

* 데이터수 == 행 수 경우 열 추가 가능
* 데이터수 == 1 경우도 열 추가 가능
* 그 외의 경우 행 추가 X
