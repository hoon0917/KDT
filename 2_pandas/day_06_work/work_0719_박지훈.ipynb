```python
# unique 메서드를 사용해 특정 열의 고유값 반환
df_book['Total_Review'].unique()
```

```python
# 특정 열의 문자열이 포함된 데이터에서 문자열과 공백, 콤마 제거하기
# 람다의 역할 : Total Review 열의 값중 문자와 콤마가 섞여있는 값들을 변경해줘야 하는데
#                 람다를 사용하지 않으면 for 반복문으로 하나씩 꺼내오는 함수를 새롭게 정의해줘야 되니까 람다 가져옴
# map()의 역할 : Total Review열의 각 값에 대해 () 내 함수 적용해줌
# str()의 역할 : replace()와 strip()은 문자열에서만 가능하니 안정적으로 이 둘을 사용할 수 있게 들어온 값을 str 타입으로 변경해줌
# replace,strip()의 역할 : 람다 함수를 통해 적용된 각각의 값들을 공백으로 바꿔주고 삭제해줌
df_book['Total_Review']=df_book['Total_Review'].map(lambda x:str(x).replace('external','').strip())
df_book['Total_Review']=df_book['Total_Review'].map(lambda x:str(x).replace(',','').strip())
df_book['Total_Review']=df_book['Total_Review'].map(lambda x:str(x).replace('review','').strip())
df_book['Total_Review']=df_book['Total_Review'].astype(float)
df_book.describe()
```

```python
# 'Rating' 열의 결측치를 fillna()활용하여 Rating열의 평균값으로 채우기
df_book['Rating'].fillna(df_book['Rating'].mean())
```

```python
# 부리길이 55 이상 행 뽑기
# 1. query() 메서드를 사용
df_peng.query('bill_length_mm>55')
```

```python
# 부리의 길이가 55 이상이면서 종이 Gentoo인 데이터 필터링 하기
df_peng.query('bill_length_mm>55 and species=="Gentoo"')
```

```python
# 부리의 길이가 55 이상이면서 종이 Gentoo인 데이터 필터링 하기
# 외부 변수 사용하기
length=55
species='Gentoo'
df_peng.query('bill_length_mm >= @length and species == @species')
```

```python
# query() 메서드에서 contains 인자를 활용하여 "oe"를 포함하는 행 추출하기
df_peng.query('island.str.contains("oe")')
```

```python
# endswith() 메서드 활용해서 species에서 'e'로 끝나는 데이터 출력하기
df_peng.query('species.str.endswith("e")')
```

```python
# species 열에서 isin을 사용해 filtering 리스트에 있는 문자열이 있는 행만 추출하기 
filtering=['Adelie','Chinstrap']
df_peng.query('species.isin(@filtering)')
```

```python
df_titan=sns.load_dataset('titanic')
df_titan
```

```python
# 각 성별의 살아남은 비율을 그룹화해서 구하기
df_titan.groupby('sex')['survived'].mean()
```

```python
# 성별과 좌석을 그룹화 해서 살아남은 비율 구하기
df_titan.groupby(['sex','pclass'])['survived'].mean()
```

```python
# 3개로 그룹화 하고 2개의 함수 결과 출력하기
df_titan.groupby(['sex','pclass','embark_town'])['survived'].agg(['mean','count'])
```

```python
# 연산 대상이 되는 열 별로 함수 다르게 사용해보기
df_titan.groupby(['sex','pclass'])[['survived','age']].agg({'survived':'mean','age':'max'})
```

```python
# apply() 메서드 사용해보기
def get_IQR(data):
    _3rd=data.quantile(.75)
    _1st=data.quantile(0.25)
    return (np.abs(_3rd-_1st)*1.5)

df_titan.groupby(['sex','pclass'])['age'].apply(get_IQR)
```
